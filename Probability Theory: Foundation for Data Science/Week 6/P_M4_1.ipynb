# Problem 2

For this problem, we're going to visualize what's happening when we go between different normal distributions.

**Part A)**

Draw at least $10000$ samples from the standard normal distribution $N(0,1)$ and store the results. Make a density histogram of these samples. Set the $x$-limits for your plot to $[-10,10]$ and your $y$-limits to $[0,0.5]$ so we can compare with the plots we'll generate in **Parts B-D**.

```r
x = rnorm(10000, mean = 0, sd = 1)
hist(x, freq = FALSE, xlim = c(-8, 8), ylim = c(0, 0.6))
```

**Part b)** Now generate $10000$ samples from a $N(2,3)$ distribution and plot a histogram of the results, with the same $x$-limits and $y$-limits. Does the histogram make sense based on the changes to parameters?

Note: Be careful with the parameters for `rnorm`. It may help to check the documentation.

```r
y = rnorm(10000, 2, sqrt(3))
hist(y, freq = FALSE, xlim = c(-8, 8), ylim = c(0, 0.6))
```

We can see that the histogram is now centered around 22, which makes sense because that is the new mean value. The histogram is also much wider, which also makes sense as we increased the variance (i.e. spread) of the data.

**Part c)**

Suppose we are only able to sample from the standard normal distribution $N(0,1)$. Could we take those samples and perform a simple transformation so that they're samples from $N(2,3)$? Try this, and plot another histogram of the transformed data, again with the same axes. Does your histogram based of the transformed data look like the histogram from **Part B**?

```r
transfromed_y = x * sqrt(3) + 2
hist(transfromed_y, freq = FALSE, xlim = c(-8, 8), ylim = c(0, 0.6))
```

We can see that the two histograms do match! The samples agree with the theory.

**Part d)** 

But can you go back the other way? Take the $N(3,4)$ samples from **Part B** and transform them into samples from $N(0,1)$? Try a few transformations and make a density histogram of your transformed data. Does it look like the plot of $N(0,1)$ data from **Part A**?

```r
y = rnorm(10000, 2, sqrt(3))
transformed_y = (y - 2) / sqrt(3)
hist(transformed_y, freq = FALSE, xlim = c(-8, 8), ylim = c(0, 0.6))
```

Again, we see that the histograms are basically the same. This should cement the idea that any normal distribution can be transformed into the standard normal, and the standard normal can be transformed to any other normal distribution.
